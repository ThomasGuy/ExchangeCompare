```python
import pandas as pd
import matplotlib.pyplot as plt
import os
%matplotlib inline
from numpy import nan
pd.set_option('precision', 9)
```

```python
def dataFormat(dataDir: str, sym: str):
    bch = pd.read_csv(f'.\\data\\{dataDir}\\{sym}btc.csv')
    bch['Timestamp'] = pd.to_datetime(bch['Timestamp'], unit='s')
    bch.set_index('Timestamp', inplace=True)
    bch.replace('na', nan, inplace=True)
    bch.dropna(axis=1, how='all', inplace=True)
    new = bch.dropna()
    new = new.astype(float)
    f = lambda x: 100*(x.max() - x.min()) / x.max()
    new['diff_%'] = new.apply(f, axis='columns')
    return new
```

```python
dataDir = 'monday'
ada = dataFormat(dataDir, 'ada')
bch = dataFormat(dataDir, 'bch')
bsv = dataFormat(dataDir, 'bsv')
btg = dataFormat(dataDir, 'btg')
dash = dataFormat(dataDir, 'dash')
eos = dataFormat(dataDir, 'eos')
etc = dataFormat(dataDir, 'etc')
eth = dataFormat(dataDir, 'eth')
iota = dataFormat(dataDir, 'iota')
ltc = dataFormat(dataDir, 'ltc')
neo = dataFormat(dataDir, 'neo')
omg = dataFormat(dataDir, 'omg')
trx = dataFormat(dataDir, 'trx')
xem = dataFormat(dataDir, 'xem')
xlm = dataFormat(dataDir, 'xlm')
xmr = dataFormat(dataDir, 'xmr')
xrp = dataFormat(dataDir, 'xrp')
xtz = dataFormat(dataDir, 'xtz')
zec = dataFormat(dataDir, 'zec')
zrx = dataFormat(dataDir, 'zrx')
```
